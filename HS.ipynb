```python
def label(row):
    if row['status'] == 'correct':             
        try:
            json_data = ast.literal_eval(row['hint'])   
            if (len(json_data['code_style']['errors'])!=0):
                return 'Correct'  
        finally:
            if pd.notna(row['hint']):
                return "Style"
            else:
                return "Correct"
    else:  
        if re.match(r"^Compilation error.*", row['hint']):
            return 'CE'
        if re.match(r"^Failed. Runtime error.*", row['hint']):
            return 'RE'        
        if re.match(r"^Failed test.*", row['hint']) or re.match(r"^Failed. Wrong answer.*", row['hint']):
            return 'WA'
        try:
            json_data = ast.literal_eval(row['hint'])
            res = ""
            if re.match(r"^Compilation error.*", json_data['message']):
                res+='CE'
            if re.match(r"^Failed test.*", json_data['message']):
                res+='WA'             
            try:
                json_data = ast.literal_eval(row['hint']) 
                if (len(json_data['code_style']['errors'])!=0):
                    return res
            finally:
                return res+"Style"
        except:
            return 'UND'
```

```python
res_list = []
def error_place(row):
    if row['solution_label'] != 'CE' :
        return ""
    else:
        list = []
        str =  row['hint']
        res = re.findall(r'[:]\d+[:]\d+[:] error[:] ', str)
        for i in res:
            d = {}
            json_data = ast.literal_eval(row['reply'])
            d['language'] = json_data['language']
            code = json_data['code']
            strings = code.splitlines()
            d['code'] = code
            
            start_message = str.find(i)
            start_num = start_message+1
            end_num =str.find(':', start_num)
            num = int(str[start_num:end_num])
            try:
                d['code_part'] = strings[num - 1]
            except:
                d['code_part'] = "Out of code"
            d['string'] = num           
            
            start_num = end_num+1
            end_num =str.find(':', start_num)
            d['symbol'] = int(str[start_num:end_num]) 
            
            start = start_message+len(i)
            end = str.find('\n',start)
            d['message'] = str[start:end]
            
            d['user_id'] = row['user_id']
            d['sub_id'] = row['submission_id']
            list.append(d)
            res_list.append(d)
        return list


df['data'] =df.apply(lambda row: error_place(row), axis=1)
df['data']
```
